```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis, QuadraticDiscriminantAnalysis
from sklearn.neural_network import MLPClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB, MultinomialNB
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from datetime import datetime
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
from sklearn.decomposition import PCA
import numpy as np

from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Flatten, Dense
from keras.utils import to_categorical

import seaborn as sns
import matplotlib.pyplot as plt
from sklearn import metrics

# Load MNIST dataset
digits = datasets.load_digits()
X = digits.images.reshape((len(digits.images), -1))
y = digits.target

# Number of iterations
n_iterations = 10

# Models to evaluate
models = {
    "Logistic Regression": LogisticRegression(multi_class='multinomial',n_jobs=-1, max_iter=10000),
    "Linear Discriminant Analysis": LinearDiscriminantAnalysis(),
    "Multi-Layer Perceptron": MLPClassifier(max_iter=10000),
    "Linear SVM": SVC(kernel='linear'),
    "Polynomial SVM (degree=2)": SVC(kernel='poly', degree=2, C=1.0),
    "Polynomial SVM (degree=3)": SVC(kernel='poly', degree=3, C=1.0),  # degree 및 C 값은 조절 가능
    "Gaussian Naive Bayes": GaussianNB(),
    "Multinomial Naive Bayes": MultinomialNB(),
    "CNN": Sequential([
        Conv2D(32, kernel_size=(4, 4), activation='relu', input_shape=(8, 8, 1)),
        MaxPooling2D(pool_size=(2, 2)),
        Flatten(),
        Dense(128, activation='relu'),
        Dense(10, activation='softmax')
    ])
}

# Dictionary to store results and confusion matrices
results = {model: {"accuracy": [], "prediction_time": [], "training_time": []} for model in models}
confusion_matrices = {model: [] for model in models}

# Iterate over models
for model_name, model in models.items():
    print(f"Training and evaluating {model_name}...")
    # Lists to store loss values for each iteration
    training_loss_per_iteration = []

    # Iterate over iterations
    for iteration in range(n_iterations):
        # CNN requires additional preprocessing for image data
        if model_name == "CNN":
            # Reshape data for CNN (assuming the original data is 8x8 images)
            X_cnn = X.reshape((-1, 8, 8, 1))

            # Split data into train and test sets
            X_train, X_test, y_train, y_test = train_test_split(X_cnn, to_categorical(y), test_size=0.2, random_state=20)

            # Compile the model
            model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
        else:
            # Split data into train and test sets
            X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=20)

        # Train the model
        start_time = datetime.now()
        if model_name == "CNN":
            history = model.fit(X_train, y_train, epochs=5, batch_size=32, verbose=1)  # 변경: verbose를 1로 설정
        else:
            model.fit(X_train, y_train)
        training_time = datetime.now() - start_time

        # Evaluate the model
        start_time = datetime.now()
        if model_name == "CNN":
            model.fit(X_train, y_train, epochs=5, batch_size=32, verbose=0)  # Additional training for CNN
            y_pred = np.argmax(model.predict(X_test), axis=1)
        else:
            y_pred = model.predict(X_test)
        prediction_time = datetime.now() - start_time

        # Convert one-hot encoding back to original labels
        if model_name == "CNN":
            y_test = np.argmax(y_test, axis=1)

        prediction_time = datetime.now() - start_time

        # Calculate accuracy
        if model_name == "CNN":
            acc = accuracy_score(y_test, y_pred)
        else:
            acc = accuracy_score(y_test, y_pred)

        # Store results
        results[model_name]["accuracy"].append(acc)
        results[model_name]["prediction_time"].append(prediction_time.total_seconds())
        results[model_name]["training_time"].append(training_time.total_seconds())

        # Confusion Matrix 계산 및 저장
        confusion_matrix = metrics.confusion_matrix(y_test, y_pred)
        confusion_matrices[model_name].append(confusion_matrix)
   # Plot the average training loss over epochs for all iterations
    if training_loss_per_iteration:  # Check if the list is not empty
        avg_loss_values = np.mean(training_loss_per_iteration, axis=0)
        plt.figure(figsize=(10, 5))
        plt.plot(range(1, len(avg_loss_values) + 1), avg_loss_values, label=f"Average Loss")

        plt.title(f"{model_name} - Average Training Loss Over Epochs")
        plt.xlabel('Epoch')
        plt.ylabel('Loss')
        plt.legend()
        plt.show()
    else:
        print(f"No training loss recorded for {model_name}")



# Calculate average results
avg_results = {}
for model_name, model_data in results.items():
    avg_results[model_name] = {
        "accuracy": sum(model_data["accuracy"]) / n_iterations,
        "prediction_time": sum(model_data["prediction_time"]) / n_iterations,
        "training_time": sum(model_data["training_time"]) / n_iterations
    }

# Print average results
print("\nAverage Results:")
for model_name, model_data in avg_results.items():
    print(f"{model_name}:")
    print(f"  Accuracy: {model_data['accuracy']:.4f}")
    print(f"  Prediction Time: {model_data['prediction_time']:.4f} seconds")
    print(f"  Training Time: {model_data['training_time']:.4f} seconds")
    print("--------------------------")
# Calculate average confusion matrix
avg_confusion_matrix = {}
for model_name, confusion_matrix_list in confusion_matrices.items():
    if confusion_matrix_list:  # Check if the list is not empty
        avg_matrix = np.mean(confusion_matrix_list, axis=0)
        # Normalize each row to make the sum of each row equal to 1
        row_sums = avg_matrix.sum(axis=1, keepdims=True)
        avg_confusion_matrix[model_name] = avg_matrix / row_sums
    else:
        avg_confusion_matrix[model_name] = None

# Visualize average confusion matrix
for model_name, avg_matrix in avg_confusion_matrix.items():
    if avg_matrix is not None:
        plt.figure(figsize=(4, 4))
        sns.heatmap(avg_matrix, annot=True, fmt=".2f", cmap="Blues", cbar=False)
        plt.title(f"Average Confusion Matrix - {model_name}")
        plt.xlabel("Predicted Label")
        plt.ylabel("True Label")
        plt.show()
    else:
        print(f"No confusion matrix available for {model_name}")

# Lists to store model names, accuracies, training times, and prediction times
model_names = []
accuracies = []
training_times = []
prediction_times = []

# Collect data for each model
for model_name, model_data in avg_results.items():
    model_names.append(model_name)
    accuracies.append(model_data['accuracy'])
    training_times.append(model_data['training_time'])
    prediction_times.append(model_data['prediction_time'])

# Plot model accuracies
plt.figure(figsize=(10, 5))
plt.bar(model_names, accuracies, color='skyblue')
plt.title('Model Accuracies')
plt.xlabel('Model')
plt.ylabel('Accuracy')
plt.ylim(0, 1)
plt.xticks(rotation='vertical')  # 세로로 작성
plt.show()

# Plot model training times
plt.figure(figsize=(10, 5))
plt.bar(model_names, training_times, color='lightcoral')
plt.title('Model Training Times')
plt.xlabel('Model')
plt.ylabel('Training Time (seconds)')
plt.xticks(rotation='vertical')  # 세로로 작성
plt.show()

# Plot model prediction times
plt.figure(figsize=(10, 5))
plt.bar(model_names, prediction_times, color='lightgreen')
plt.title('Model Prediction Times')
plt.xlabel('Model')
plt.ylabel('Prediction Time (seconds)')
plt.xticks(rotation='vertical')  # 세로로 작성
plt.show()
```
